```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder

# Iterate over each dataset
for language, dataset_info in datasets.items():
    print(f"Processing {language} dataset...")

    # Split the data into train and test sets
    X_train_text, X_test_text, y_train, y_test = train_test_split(dataset_info['data']['preprocessed_text'], dataset_info['data']['outcome'], test_size=0.2, random_state=42)

    # Vectorize the text data using TF-IDF
    vectorizer = TfidfVectorizer()
    X_train = vectorizer.fit_transform(X_train_text)
    X_test = vectorizer.transform(X_test_text)

    # Logistic Regression
    lr_model = LogisticRegression()
    lr_model.fit(X_train, y_train)
    lr_preds = lr_model.predict(X_test)
    print("Logistic Regression:")
    print(classification_report(y_test, lr_preds))

    # Support Vector Machine
    svm_model = SVC()
    svm_model.fit(X_train, y_train)
    svm_preds = svm_model.predict(X_test)
    print("Support Vector Machine:")
    print(classification_report(y_test, svm_preds))

    # Random Forest
    rf_model = RandomForestClassifier()
    rf_model.fit(X_train, y_train)
    rf_preds = rf_model.predict(X_test)
    print("Random Forest:")
    print(classification_report(y_test, rf_preds))

    # Encode target classes into integer labels
    label_encoder = LabelEncoder()
    y_train_encoded = label_encoder.fit_transform(y_train)
    y_test_encoded = label_encoder.transform(y_test)

    # XGBoost
    xgb_model = XGBClassifier()
    xgb_model.fit(X_train, y_train_encoded)
    xgb_preds = xgb_model.predict(X_test)
    print("XGBoost:")
    print(classification_report(y_test_encoded, xgb_preds, target_names=label_encoder.classes_))

    # Simple Neural Network
    mlp_model = MLPClassifier()
    mlp_model.fit(X_train, y_train)
    mlp_preds = mlp_model.predict(X_test)
    print("Simple Neural Network:")
    print(classification_report(y_test, mlp_preds))

```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from emoji import demojize

# Assuming datasets is a dictionary containing the datasets for each language
datasets = {
    'english': {'data': english_df, 'outcome_column': 'outcome'},
    'hindi': {'data': hindi_df, 'outcome_column': 'outcome'},
    'german': {'data': german_df, 'outcome_column': 'outcome'}
}


# Initialize models
models = {
    'Logistic Regression': LogisticRegression(),
    'Support Vector Machine': SVC(),
    'Random Forest': RandomForestClassifier(),
    'XGBoost': XGBClassifier(),
    'Simple Neural Network': MLPClassifier()
}

# Download necessary NLTK resources if not already downloaded
nltk.download('punkt')

# Define function for minimal preprocessing
def minimal_preprocess(text):
    # Substitute "RT @user" with an empty string
    text = re.sub(r'RT\s+@\w+\s*:', '', text)
    # Substitute user mentions with @USER and remove @USER
    text = re.sub(r'@\w+', '', text)
    # Substitute URLs with URL and remove URL
    text = re.sub(r'http\S+', '', text)
    # Replace emojis with text representation
    text = demojize(text)
    # Remove hashtag symbol from hashtags
    text = re.sub(r'#(\w+)', r'\1', text)
    # Replace extra blank spaces with a single space
    text = re.sub(r'\s+', ' ', text)
    return text.strip()  # Strip any leading/trailing spaces

# Process each dataset
for language, data_info in datasets.items():
    print(f"Processing {language} dataset...")

    # Load dataset
    df = data_info['data']
    outcome_column = data_info['outcome_column']

    # Preprocess text data
    df['preprocessed_text'] = df['text'].apply(minimal_preprocess)

    # Split data into train, validation, and test sets
    X_train_text, X_test_text, y_train, y_test = train_test_split(df['preprocessed_text'], df[outcome_column], test_size=0.2, random_state=42)
    X_train_text, X_val_text, y_train, y_val = train_test_split(X_train_text, y_train, test_size=0.2, random_state=42)

    # Vectorize text data using TF-IDF
    vectorizer = TfidfVectorizer()
    X_train = vectorizer.fit_transform(X_train_text)
    X_val = vectorizer.transform(X_val_text)
    X_test = vectorizer.transform(X_test_text)

    # Encode target classes into integer labels
    label_encoder = LabelEncoder()
    y_train_encoded = label_encoder.fit_transform(y_train)
    y_val_encoded = label_encoder.transform(y_val)
    y_test_encoded = label_encoder.transform(y_test)

    # Train and evaluate models
    model_reports = {}
    for model_name, model in models.items():
        print(f"Training and evaluating {model_name}...")
        model.fit(X_train, y_train_encoded)
        val_preds = model.predict(X_val)
        test_preds = model.predict(X_test)
        model_reports[model_name] = {
            'Validation': {'y_true': y_val_encoded, 'y_pred': val_preds},
            'Test': {'y_true': y_test_encoded, 'y_pred': test_preds}
        }

    # Print classification reports for each model and dataset
    print(f"\nClassification reports for {language} dataset:")
    for model_name, reports in model_reports.items():
        print(f"\n{model_name} Model:")
        print("Validation Set:")
        print(classification_report(reports['Validation']['y_true'], reports['Validation']['y_pred'], target_names=label_encoder.classes_))
        print("\nTest Set:")
        print(classification_report(reports['Test']['y_true'], reports['Test']['y_pred'], target_names=label_encoder.classes_))
```

```python
from sklearn.preprocessing import LabelEncoder

# Define datasets for different languages
datasets = {
    'english': {'data': english_df, 'outcome_column': 'outcome'},
    'hindi': {'data': hindi_df, 'outcome_column': 'outcome'},
    'german': {'data': german_df, 'outcome_column': 'outcome'}
}
# Initialize best model variables
best_model = None
best_accuracy = 0

# Iterate over each dataset
for lang, dataset_info in datasets.items():
    data = dataset_info['data']
    outcome_column = dataset_info['outcome_column']

    # Split the data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(data['preprocessed_text'], data[outcome_column], test_size=0.2, random_state=42)

    # Vectorize the text data using TF-IDF
    vectorizer = TfidfVectorizer()
    X_train_tfidf = vectorizer.fit_transform(X_train)
    X_test_tfidf = vectorizer.transform(X_test)

    # Encode target classes into integer labels
    label_encoder = LabelEncoder()
    y_train_encoded = label_encoder.fit_transform(y_train)
    y_test_encoded = label_encoder.transform(y_test)

    # Train and evaluate models
    models = {
        'Logistic Regression': LogisticRegression(),
        'Support Vector Machine': SVC(),
        'Random Forest': RandomForestClassifier(),
        'XGBoost': XGBClassifier(),
        'Simple Neural Network': MLPClassifier()
    }

    for model_name, model in models.items():
        model.fit(X_train_tfidf, y_train_encoded)
        accuracy = model.score(X_test_tfidf, y_test_encoded)
        print(f"{lang} - {model_name} Accuracy: {accuracy}")

        # Update best model if current model has higher accuracy
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_model = f"{lang} - {model_name}"

print(f"\nBest Model: {best_model} with Accuracy: {best_accuracy}")

```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Define datasets for different languages
datasets = {
    'english': {'data': english_df, 'outcome_column': 'outcome'},
    'hindi': {'data': hindi_df, 'outcome_column': 'outcome'},
    'german': {'data': german_df, 'outcome_column': 'outcome'}
}

# Iterate over each dataset
for lang, dataset_info in datasets.items():
    data = dataset_info['data']
    outcome_column = dataset_info['outcome_column']

    # Split the data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(data['preprocessed_text'], data[outcome_column], test_size=0.2, random_state=42)

    # Vectorize the text data using TF-IDF
    vectorizer = TfidfVectorizer()
    X_train_tfidf = vectorizer.fit_transform(X_train)
    X_test_tfidf = vectorizer.transform(X_test)

    # Initialize Decision Tree classifier
    clf = DecisionTreeClassifier()

    # Train the model
    clf.fit(X_train_tfidf, y_train)

    # Make predictions on the test set
    y_pred = clf.predict(X_test_tfidf)

    # Print classification report
    print(f"Classification Report for {lang} Dataset:")
    print(classification_report(y_test, y_pred))
```
